# Task 1: Problem Definition and Dataset Exploration
## Cardiovascular Disease Prediction

**Problem Statement:** Predict whether a patient has cardiovascular disease based on medical examination data including age, gender, blood pressure, cholesterol, glucose levels, and lifestyle factors.

**Dataset:** [Cardiovascular Disease Dataset](https://www.kaggle.com/datasets/sulianova/cardiovascular-disease-dataset) containing 70,000 patient records with 12 features.

**Target Variable:** `cardio` (1 = has cardiovascular disease, 0 = no cardiovascular disease)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', palette='viridis')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
df = pd.read_csv('cardio_train.csv', sep=';')
print(f"Dataset Shape: {df.shape}")
print(f"Total Records: {df.shape[0]:,}")
print(f"Total Features: {df.shape[1]}")
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
print("Missing Values:")
print(df.isnull().sum())
print(f"\nDuplicate Rows: {df.duplicated().sum()}")
```

```python
df['age_years'] = (df['age'] / 365.25).astype(int)
print(f"Age Range: {df['age_years'].min()} - {df['age_years'].max()} years")
print(f"\nTarget Distribution:")
print(df['cardio'].value_counts())
print(f"\nTarget Balance: {df['cardio'].value_counts(normalize=True).round(4) * 100}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

df['cardio'].value_counts().plot(kind='bar', ax=axes[0], color=['#2ecc71', '#e74c3c'])
axes[0].set_title('Target Variable Distribution', fontweight='bold')
axes[0].set_xticklabels(['No Disease (0)', 'Disease (1)'], rotation=0)
axes[0].set_ylabel('Count')

axes[1].pie(df['cardio'].value_counts(), labels=['No Disease', 'Disease'],
            autopct='%1.1f%%', colors=['#2ecc71', '#e74c3c'], startangle=90,
            explode=(0.05, 0.05))
axes[1].set_title('Target Variable Proportion', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
numerical_cols = ['age_years', 'height', 'weight', 'ap_hi', 'ap_lo']

for idx, col in enumerate(numerical_cols):
    row, c = idx // 3, idx % 3
    df[col].hist(bins=50, ax=axes[row][c], color='#3498db', edgecolor='black', alpha=0.7)
    axes[row][c].set_title(f'Distribution of {col}', fontweight='bold')
    axes[row][c].set_xlabel(col)

axes[1][2].axis('off')
plt.tight_layout()
plt.show()
```

```python
categorical_cols = ['gender', 'cholesterol', 'gluc', 'smoke', 'alco', 'active']
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

for idx, col in enumerate(categorical_cols):
    row, c = idx // 3, idx % 3
    df[col].value_counts().plot(kind='bar', ax=axes[row][c], color=sns.color_palette('Set2'))
    axes[row][c].set_title(f'{col} Distribution', fontweight='bold')
    axes[row][c].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

## Initial Observations

1. **Balanced Dataset** — Target variable is nearly balanced (~50/50 split)
2. **Age in Days** — Age is recorded in days, needs conversion to years for interpretability
3. **Outliers Present** — Blood pressure columns (`ap_hi`, `ap_lo`) contain extreme outliers (negative values, values > 1000)
4. **Categorical Features** — `cholesterol` and `gluc` are ordinal (1=normal, 2=above normal, 3=well above normal)
5. **Binary Features** — `smoke`, `alco`, `active`, `gender` are binary
6. **No Missing Values** — Dataset has no null values
7. **ID Column** — `id` column is not useful for prediction and should be dropped
