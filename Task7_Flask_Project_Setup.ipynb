## 7.3 Key Flask Concepts

| Concept | Description |
|---------|-------------|
| **Routes** | URL endpoints mapped to Python functions |
| **Templates** | HTML files rendered using Jinja2 |
| **Static Files** | CSS, JS, images served from `/static` |
| **Request** | Access form data, JSON, query params |
| **jsonify** | Return JSON responses from endpoints |
| **render_template** | Render HTML templates with dynamic data |

````python
import os

os.makedirs('templates', exist_ok=True)
os.makedirs('static/css', exist_ok=True)
os.makedirs('static/js', exist_ok=True)

print("Flask project directories created:")
for root, dirs, files in os.walk('.'):
    level = root.replace('.', '').count(os.sep)
    indent = ' ' * 2 * level
    subdir = os.path.basename(root) or '.'
    if any(skip in root for skip in ['.ipynb', '__pycache__', '.git']):
        continue
    print(f'{indent}{subdir}/')
````
