## 10.1 Verify All Files

````python
import os

all_files = {
    'Dataset': ['cardio_train.csv', 'cardio_cleaned.csv'],
    'Models': [
        'model_linear_regression.pkl', 'model_logistic_regression.pkl',
        'model_svm.pkl', 'model_knn.pkl',
        'model_decision_tree.pkl', 'model_random_forest.pkl', 'scaler.pkl'
    ],
    'Flask': ['app.py', 'templates/index.html', 'static/css/style.css', 'static/js/app.js'],
    'Notebooks': [
        'Task1_Problem_Definition_and_Dataset_Exploration.ipynb',
        'Task2_Data_Cleaning_and_Preprocessing.ipynb',
        'Task3_Model_Creation.ipynb',
        'Task4_Model_Evaluation.ipynb',
        'Task5_Advanced_Model_Training.ipynb',
        'Task6_Visualization_of_Metrics.ipynb',
        'Task7_Flask_Project_Setup.ipynb',
        'Task8_Create_Frontend.ipynb',
        'Task9_Create_Backend.ipynb',
        'Task10_Deployment.ipynb'
    ]
}

for category, files in all_files.items():
    print(f"\n{category}:")
    for f in files:
        exists = os.path.exists(f)
        size = os.path.getsize(f) if exists else 0
        status = f'✓ ({size:,} bytes)' if exists else '✗ MISSING'
        print(f"  {status:>20s}  {f}")
````

## 10.2 Requirements

````python
requirements = """flask
pandas
numpy
matplotlib
seaborn
scikit-learn
plotly
joblib
"""

with open('requirements.txt', 'w') as f:
    f.write(requirements)

print("requirements.txt created.")
print(requirements)
````

## 10.3 Run the Application

### Local Deployment

```bash

python app.py
```
